# SERIES DE TIEMPO PREDICCIONES

Librerias necesarias

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)#Verificamos versión vigente
```

Agrupación de la serie de tiempo:

```python
def Agrupacion(time_series):
  time_serie=time_series.groupby(pd.Grouper(freq='1M')).sum()#Agrupación por mes
  return time_serie
```

Partición de los datos:

```python
def Particion(data):
  split_time=int(len(data)*0.8)
  time=data.index.to_numpy()
  series=data['CANTFACTURADA'].to_numpy()
  time_train = time[:split_time]
  x_train = series[:split_time]
  time_valid = time[split_time:]
  x_valid = series[split_time:]
  return time_train, x_train, time_valid, x_valid, series, split_time, time
```

Función para graficar:

```python
def plot_series(time, series, format="-", start=0, end=None):
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid(True)
```

Función para configuración de series de tiempo:

En la siguiente función, se prepara el dataset para generar lotes de ventanas de características y etiquetas barajadas, para llevarlas al modelo:

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
  dataset = tf.data.Dataset.from_tensor_slices(series)#Retorna los elemtos de la serie segmentados
  dataset = dataset.window(window_size + 1, shift=1, drop_remainder=True)#Corta los datos en el tamaño de la ventana, shift=1 es el deplazamiento de 1 por ventana y drop_remainder, corta los datos, para solo tener windows size iguales
  dataset = dataset.flat_map(lambda window: window.batch(window_size + 1))#se ponen los datos en listas
  dataset = dataset.shuffle(shuffle_buffer).map(lambda window: (window[:-1], window[-1]))#Se dividen los datos en entidades y etiquetas (inputs and labels), inputs=todos los datos, labels=último dato. El shuffle reorganiza los tiempos
  dataset = dataset.batch(batch_size).prefetch(1)#se escogen los lotes de datos agrupando de acuerdo al batch size
  return dataset
```

Función de predicción (forescast_funtion)

Nos entrega la comparación entre la predicción y los datos de test a través de un gráfico

##Estacionalidad de las piezas

Para verificar si una serie es estacionario, se realiza la prueba de Dickey- Fuller Aumentada. En esta prueba, la hipótesis nula es que una serie de tiempo no es estacionaria. Es decir,



*   p-value > 0.05, no se puede rechazar la hipótesis nula y los datos NO SON ESTACIONARIOS.
*   p-value < 0.05, se rechaza la hipótesis nula y los datos SON ESTACIONARIOS.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def Verificacion_estacional(time_series):
  resultado=adfuller(time_series)
  labels=['   Estadístico ADF','   p-value','   # de retrasos','   Observación']
  for valor,label in zip(resultado,labels):
    print(label + ':' +str(valor))
  if resultado[1]<=0.05:
    print('   Se rechaza la hipótesis nula y los datos SON ESTACIONARIOS')
  else:
    print('   No se rechaza la hipótesis nula y los datos NO SON ESTACIONARIOS')
```

```python
Datas=[dataP0,dataP1,dataP2,dataP3,dataP4,dataP5]
Productos_PIEZAS = ['PT01025A', 'PT02025R', 'PT02025I', 'PT59202I', 'PT59202A', 'PT01002A']
for i in range(len(Datas)):
  print(f'La pieza {Productos_PIEZAS[i]} en la prueba de Dickey- Fuller Aumentada prensenta los siguientes resultados:')
  Verificacion_estacional(Datas[i]['CANTFACTURADA'])
```
